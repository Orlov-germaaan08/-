```python
import numpy as np
import matplotlib.pyplot as plt

res = []
x0 = 5.0 
lambda_ = 0.1  

def f(x):
    return 3*(x)**2 

def df(x):
    return 6*x 

while True:
    res.append([x0, f(x0)])
    x1 = x0 - lambda_ * df(x0)
    if np.linalg.norm(x0 - x1) <= 0.00000001:
        break
    x0 = x1
```

```python
res = np.array(res)
plt.plot(res[:, 0], res[:, 1], 'ro-')
plt.plot()
x = np.linspace(-10, 10, 50)  
y=f(x)
plt.xlim(-5, 5) 
plt.ylim(0, 100)
plt.plot(x, y)
plt.show()
```

```python
def f_x(x, y):
    return 3*x**2 + 5*y**2
```

```python
def df(x, y):
    return np.array([6 * x, 10*y])

x0 = np.array([5, 10])
lam = 0.1
res = []
x1 = x0 - lam*df(*x0)

while np.linalg.norm(x1 - x0) > 0.00000001:
    res.append([x0[0], x0[1]])
    x0 = x1
    x1 = x0 - lam*df(*x0)
```

```python
res = np.array(res)

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = f_x(x, y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, alpha=0.5)
ax.plot(res[:,0], res[:,1], f_x(res[:,0], res[:,1]), color='r', marker='o')
plt.show()
```
